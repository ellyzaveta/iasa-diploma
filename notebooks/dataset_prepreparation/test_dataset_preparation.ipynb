```python
import pandas as pd
```

```python
df_test = pd.read_csv('../test_data.csv')
```

```python
df_test_inclusion = pd.DataFrame()

df_test_inclusion[['query_id', 'patient_note', 'doc_id', 'trial_title',
    'trial_criteria', 'trial_criteria_total_number', 'relevance']] = (
    df_test)[['query_id', 'text', 'doc_id', 'title', 'inclusion_criteria', 'inclusion_criteria_number', 'relevance']]

df_test_inclusion['trial_criteria_type'] = 'inclusion'

df_test_inclusion = df_test_inclusion[['query_id', 'patient_note', 'doc_id', 'trial_title',
         'trial_criteria', 'trial_criteria_total_number', 'trial_criteria_type', 'relevance']]
```

```python
df_test_exclusion = pd.DataFrame()

df_test_exclusion[['query_id', 'patient_note', 'doc_id', 'trial_title',
    'trial_criteria', 'trial_criteria_total_number', 'relevance']] = (
    df_test)[['query_id', 'text', 'doc_id', 'title', 'exclusion_criteria', 'exclusion_criteria_number', 'relevance']]

df_test_exclusion['trial_criteria_type'] = 'exclusion'

df_test_exclusion = df_test_exclusion[['query_id', 'patient_note', 'doc_id', 'trial_title',
         'trial_criteria', 'trial_criteria_total_number', 'trial_criteria_type', 'relevance']]
```

```python
import matplotlib.pyplot as plt

import matplotlib as mpl
import styles.plt_style as style
import styles.colors as colors

mpl.rcParams.update(mpl.rcParamsDefault)
plt.style.use(style.plt_style)

mpl.rcParams['font.size'] = style.plt_font_size
mpl.rcParams['font.family'] = style.plt_font_style
```

```python
def plot_value_counts(dataframe, column='query_id'):

    value_counts = dataframe[column].value_counts()

    plt.figure(figsize=(30, 6))
    value_counts.plot(kind='bar', color=colors.orange_base)
    plt.xlabel(column)
    plt.ylabel('Counts')
    plt.title(f'Value Counts for {column}')
    plt.show()
```

```python
plot_value_counts(df_test_inclusion)
```

```python
plot_value_counts(df_test_exclusion)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_test_inclusion_full, df_test_inclusion_sampled = train_test_split(df_test_inclusion, test_size=0.15, random_state=42, stratify=df_test_inclusion['query_id'])
```

```python
plot_value_counts(df_test_inclusion_sampled)
```

```python
df_test_inclusion_sampled.shape
```

```python
df_test_exclusion_full, df_test_exclusion_sampled = train_test_split(df_test_exclusion, test_size=0.15, random_state=42, stratify=df_test_exclusion['query_id'])
```

```python
plot_value_counts(df_test_exclusion_sampled)
```

```python
df_test_exclusion_sampled.shape
```

```python
def plot_length_distribution(df_input):
    dataframe = pd.DataFrame(df_input)

    dataframe['text_length'] = dataframe[['patient_note', 'trial_title', 'trial_criteria']].apply(lambda x: x.str.len().sum(), axis=1)
    
    dataframe['length_bin'] = pd.cut(dataframe['text_length'], bins=4)
    
    hist_data = dataframe['length_bin'].value_counts().sort_index()
    
    colors_arr = [colors.orange_base, colors.orange_2, colors.orange_3, colors.orange_4]

    plt.figure(figsize=(4, 4)) 
    patches, texts, autotexts = plt.pie(hist_data, 
                                        autopct='%1.1f%%', 
                                        colors=colors_arr,
                                        startangle=140,
                                        wedgeprops={'edgecolor': 'white', 'linewidth': 1.5}) 

    plt.title('Length Distribution')
    
    plt.legend(patches, hist_data.index.astype(str), loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    
    plt.show()
```

```python
plot_length_distribution(df_test_inclusion_sampled)
```

```python
plot_length_distribution(df_test_exclusion_sampled)
```

```python
df_test_merged = pd.concat([df_test_inclusion_sampled, df_test_exclusion_sampled], axis=0)
```
